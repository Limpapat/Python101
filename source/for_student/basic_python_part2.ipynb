## Condition & Loop

### If...else condition

```
if Boolean-expression:
  if_block_statement
else:
  else_block_statement
```

### While loop

```
while Boolean-expression:
  block statement
```

### For loop

```
for iterate_var in sequence:
  statement
```

### Loop controller : break, continue & else

**break**
```
while/for loop:
  statement inside loop
  if condition:
    break -----------------|
  statement inside loop    |
statement outside loop  <--|
```

**continue**
```
while/for loop:
  statement inside loop <--|
  if condition:            |
    continue --------------|
  statement inside loop
statement outside loop
```

**else**
```
while/for loop:
  statement
  if condition: break
  if condition: continue
else:
  statement
statement outside loop
```
