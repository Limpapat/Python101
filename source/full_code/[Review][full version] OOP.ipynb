### Multiple inheritance

```
# Class ParentA :
#   Member_ParentA
#    ........
# Class ParentB :
#   Member_ParentB
#    ........
# Class Child (ParentA, ParentB):
#   Member_Child
#    ........
```

````
class father :
  def __init__(self):
    print("This is Darth Vader, I'm your father!")
class mother :
  def __init__(self):
    print("This is Padme, I'm your mother!")
class child(father, mother) :
  def __init__(self):
    self.name = "Luke Skywalker"
  def print(self):
    father.__init__(self)
    mother.__init__(self)
    print("This is", self.name, ", I'm jedi!")
````

````
C = child()
C.print()
````

````
class Rectangle :
  def __init__(self, width, length):
    self.width = width
    self.length = length
  def area(self):
    return self.width * self.length
  def perimeter(self):
    return 2 * self.length + 2 * self.width

class Square :
  def __init__(self, length):
    self.length = length
  def area(self):
    return self.length * self.length
  def perimeter(self):
    return 4 * self.length
````

#### Parameterless super( )

````
class Square(Rectangle):
  def __init__(self, length):
    super().__init__(length, length)

square = Square(4)
square.area(), square.perimeter()
````

#### Super( ) with parameters

```
# super(subclass, obj)
```



````
class Square(Rectangle):
  def __init__(self, length):
    super(Square, self).__init__(length, length)

square = Square(4)
square.area()
````

````
class Cube(Square):
  def surface_area(self):
    face_area = super(Square, self).area()
    return 6 * face_area
  def volumn(self):
    face_area = super(Square, self).area()
    return face_area * self.length

cube = Cube(4)
cube.surface_area()
````

````
class SomeBaseClass(object):
  def __init__(self):
    print('SomeBaseClass.__init__(self) called')

class UnsuperChild(SomeBaseClass):
  def __init__(self):
    print('UnsuperChild.__init__(self) called')
    SomeBaseClass.__init__(self)

class SuperChild(SomeBaseClass):
  def __init__(self):
    print('SuperChild.__init__(self) called')
    super(SuperChild, self).__init__()
````

````
class InjectMe(SomeBaseClass):
  def __init__(self):
    print('InjectMe.__init__(self) called')
    super(InjectMe, self).__init__()

class UnsuperInjector(UnsuperChild, InjectMe): pass

class SuperInjector(SuperChild, InjectMe): pass
````

````
o = UnsuperInjector()
````

````
o2 = SuperInjector()
````

````
### Example ###
class Triangle:
  def __init__(self, base, height):
    self.base = base
    self.height = height
  def tri_area(self):
    return 0.5 * self.base * self.height
    
class RightPyramid(Square, Triangle):
  'Pyramid with a square base'
  def __init__(self, base, slant_height):
    self.base = base
    self.slant_height = slant_height
    super().__init__(self.base)
  def area(self):
    base_area = super().area()
    perimeter = super().perimeter()
    return 0.5 * perimeter * self.slant_height + base_area
  def area_2(self):
    base_area = super().area()
    triangle_area = super().tri_area()
    return 4 * triangle_area + base_area
````

````
pyramid = RightPyramid(4, 2)

# get error
pyramid.area_2()
````

Hint: you need to do two things to fix this,
* 1. All methods that are called with super( ) need to have a call to their superclass’s version of that method. This means that you will need to add super( ).\_\_init\_\_( ) to the .\_\_init\_\_( ) methods of Triangle and Rectangle.
* 2. Redesign all the .\_\_init\_\_( ) calls to take a keyword dictionary.

````
# Complete code
class Rectangle :
  def __init__(self, width, length, **kwargs):
    self.width = width
    self.length = length
    print('Reg', kwargs)
    super().__init__(**kwargs)
  def area(self):
    return self.width * self.length
  def perimeter(self):
    return 2 * self.length + 2 * self.width

class Square(Rectangle):
  def __init__(self, length, **kwargs):
    print('Square', kwargs)
    super().__init__(length=length, width=length, **kwargs)

class Cube(Square):
  def surface_area(self):
    face_area = super().area()
    return 6 * face_area
  def volumn(self):
    face_area = super().area()
    return face_area * self.length

class Triangle:
  def __init__(self, base, height, **kwargs):
    self.base = base
    self.height = height
    print('Triangle', kwargs)
    super().__init__(**kwargs)
  def tri_area(self):
    return 0.5 * self.base * self.height

class RightPyramid(Triangle, Square):
  'Pyramid with a square base'
  def __init__(self, base, slant_height, **kwargs):
    self.base = base
    self.slant_height = slant_height
    kwargs["height"] = slant_height
    kwargs["length"] = base
    kwargs["base"] = base
    print('RP', kwargs)
    super().__init__(**kwargs)
  def area(self):
    base_area = super().area()
    perimeter = super().perimeter()
    print('base area', base_area, 'perimeter', perimeter)
    return 0.5 * perimeter * self.slant_height + base_area
  def area_2(self):
    base_area = super().area()
    triangle_area = super().tri_area()
    print('base area', base_area, 'tri_area', triangle_area)
    return 4 * triangle_area + base_area
````

````
pyramid = RightPyramid(base=2,slant_height=3)
print(RightPyramid.__mro__)
print(pyramid.area(), pyramid.area_2())
````

## Encapsulation

```
# public
# __private    -> called dunder -> only in class
# _protected   -> called under  -> in class and subclass
```



````
class classA :
  public = "I'm public."
  __private = "I'm private."
  _protected = "I'm protected."
  def __init__(self):
    print("---Inside Class A---")
    print(self.public)
    print(self.__private)
    print(self._protected)
  def printA(self):
    print("Private data is...", self.__private)
class subclassB(classA) :
  def __init__(self):
    print("---Inside Subclass B---")
    print(self.public)
    print(self._protected)
````

````
a = classA()
b = subclassB()

# public
print("---Outside Class A")
print(a.public)
print("---Outside Class B")
print(b.public)
````

````
# protected
print(b._protected)
print(a._protected)
````
