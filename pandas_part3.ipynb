```python
gb = df['nationality'].groupby(df['nationality']).count()
```

```python
nationality, number = [], []
for i,j in gb.iteritems():
  nationality.append(i)
  number.append(j)

df_nation = pd.DataFrame({'nationality':nationality, 'number':number})
pd.set_option('display.max_rows', df_nation.shape[0]+1)
df_nation
```

```python
max = df_nation['number'].max()
df_nation[df_nation['number']==max]
```
